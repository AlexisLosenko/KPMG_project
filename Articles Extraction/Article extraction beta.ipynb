```python
from pymongo import MongoClient
```

```python
client = MongoClient()
db = client.kpmg
stat_coll = db.statutes
```

```python
uid = "0686873232"
stat_doc = stat_coll.find_one({"_id": uid})
doc_data = stat_doc['documents']
date = [i for i in doc_data][0]
text = next(iter(next(iter(doc_data.values())).values()))
```

```python
object = doc_data[date]['object']
constitution = ['CONSTITUTION', 'OPRICHTING']
for word in object.split():
    if word in constitution:
        print('yo')
```

```python
split = text.split('\n')
```

```python
count=0
for i in split:
    count += 1
    print(f"{count}/{i}")
```

```python
key_word = ["Art", "Article", 'Art.', 'Art,', 'Ast', 'Ari', "art" ,"At", "ARTICLE", "ART", 'Ant', 'Artikel', 'ARTIKEL', "article"]
```

```python
split
```

```python
cut_index = []
for group in split:
    for word in group.translate({ord(i): ' ' for i in ":[‘'(),. "}).split():    
        if word in key_word:
            #print(split.index(group))
            print(group)
            cut_index.append(split.index(group))
print(cut_index)            
```

```python
for i in cut_index:
    print(i , (cut_index[cut_index.index(i)+1]))
```

```python
#Articles = [(split[i:cut_index[cut_index.index(i)+1]] for i in cut_index if i != cut_index[-1])]
```

```python
start = split[:cut_index[0]-1]

articles = []
for i in cut_index:
    if i != cut_index[-1]:
        articles.append(split[i:cut_index[cut_index.index(i)+1]])
        
end = split[cut_index[-1]+1:]

sections = [start]+articles+[end]
```

```python
for element in sections:
    temp = str()
    for el in element:
        temp += el
        
    sections[sections.index(element)] = temp
```

```python
stat_coll.update_one({'_id': uid},
                                     {'$addToSet': {
                                         f"documents.{date}.sections": {str(sections.index(element)) : element for element in sections}
                                     }},
                                     upsert=True
                                     )
```
