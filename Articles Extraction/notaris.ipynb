```python
from pymongo import MongoClient
import json

client = MongoClient()
db = client.kpmg
stat_coll = db.statutes

filename = r'C:\Users\sebch\Desktop\kpmg\GitHub Repo\KPMG_project\scraping_final\scraping_staatblad_seb\link_pdf_2018.json'
with open(filename, 'r') as f:
    jason = json.load(f)
    
date_list = []
for key, value in jason.items():
    date_list.append(value['source_date'])   
date_set = set(date_list)
date_list = list(date_set)

      
```

```python
dates = date_list 
key_word = ['notaris', 'notaire']

#dates = ["20190903"]
for date in dates:
    cursor = stat_coll.find({f'documents.{date}.object': {"$regex" : "OPRICHTING | EINDE | CONSTITUTION | STATUTS | STATUTEN"}})
    for i in cursor:
        print(i["_id"])
        doc_data = i['documents']
        text = next(iter(next(iter(doc_data.values())).values()))
        #print(text)
        
        
        indices = [i for i, x in enumerate(text.split()) if x in key_word]
        print(indices)
```
