```python
#On commence par extraire les liens des sites web voulus en les sauvant en json
import scrapy
import json
import re

#Pour extraire depuis un site web on utilise le framework Scrapy.
#Il faut cr√©er la classe telle quelle. 
# POur lancer le fichier on utilise la commande:
#      scrapy runspider myspider.py

class BlogSpider(scrapy.Spider):
    name = 'DE STAATSBLADMONITOR '
    start_urls = []
    year = 2019

    for month in range(1, 9):
        month = f'0{month}'
        for day in range(1, 31):
            if day < 10:
                day = f'0{day}'
            start_urls.append(f'https://www.staatsbladmonitor.be/oprichtingen-bedrijven.html?datum={year}-{month}-{day}')

    data_final = []
    data_final_clean = []

    def parse(self, response):
        
        for data in response.css('.data::text'):
            if re.match('^[0-9]+$', data.get()):
                self.data_final.append('https://www.staatsbladmonitor.be/bedrijfsfiche.html?ondernemingsnummer='+data.get())

        with open('link_each_entreprise.json', 'w', encoding='utf-8') as f:
            json.dump(self.data_final, f, ensure_ascii=False, indent=4)
```

```python
#On importe les liens contenus dans le ficher json avec pandas
import json
import pandas as pd 

df=pd.read_json(r'link_each_entreprise.json')
print(df)
```

```python
#On change la df en liste
from urllib.request import urlopen
from bs4 import BeautifulSoup

for index, url in df.head(300).items():
    df1=list(url)
   
#Avec BeautifulSoup on extrait les liens pdf pour les sauver dans un fichier json

annuaire_pdf={}

for url in df1:
    vat_number = url[-10:]
    html_page = urlopen(url)
    soup = BeautifulSoup(html_page)
    for link in soup.findAll('a', attrs={'href': re.compile("pdf$")}):
        link_pdf= link.get('href')
        if not "static" in link_pdf :
            annuaire_pdf[vat_number] = link_pdf
            print(vat_number)
            print(link_pdf)
          
        
        
```
