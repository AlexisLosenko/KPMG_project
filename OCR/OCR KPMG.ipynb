```python
try:
    from PIL import Image
except ImportError:
    import Image
import pytesseract



def ocr_core(filename):
    # We'll use Pillow's Image class to open the image and pytesseract to detect the string in the image
    nom_fichier=str(filename[30:-4]) 

    text = pytesseract.image_to_string(Image.open(filename))  

    with open('C:\\Users\\Pierro\\Documents\\txt\\{}.txt'.format(nom_fichier), 'w') as f:  
        f.write(text)
    
    return text
    

print(ocr_core(r'C:\Users\Pierro\Documents\img\0715900580-0.jpg'))
```
